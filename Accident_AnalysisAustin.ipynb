```python
import pandas as pd

# Loading the data
data = pd.read_csv("Austin_cleaned_crash_data.csv")

# Displaying the first few rows of the dataset
data.head(10)
```

# New Section

```python
# Displaying basic statistics for the dataset
data.describe()
```

```python
# Selecting numerical columns only
numerical_data = data.select_dtypes(include=['float64', 'int64'])

# Calculating the correlations
correlations = numerical_data.corr()

print(correlations)
```

```python
# Selecting numerical columns only
numerical_data = data.select_dtypes(include=['float64', 'int64'])

# Calculating the standard deviations for the numerical columns
std_devs = numerical_data.std()

# Displaying the standard deviations
print(std_devs)
```

```python

# Calculate means for the numerical columns
means = numerical_data.mean()

# Display the means
print(means)
```

```python
# Calculate medians for the numerical columns
medians = numerical_data.median()

# Display the medians
print(medians)
```

```python
import matplotlib.pyplot as plt
# Scatterplot for Latitude vs Longitude
plt.scatter(data['rpt_latitude'], data['rpt_longitude'])
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Scatterplot of Latitude vs Longitude')
plt.show()
```

```python
# Scatterplot for Crash Severity vs Suspected Serious Injury Count
plt.scatter(data['crash_sev_id'], data['sus_serious_injry_cnt'])
plt.xlabel('Crash Severity')
plt.ylabel('Suspected Serious Injury Count')
plt.title('Scatterplot of Crash Severity vs Suspected Serious Injury Count')
plt.show()
```

```python

# Convert the 'crash_date' column to datetime format
data['crash_date'] = pd.to_datetime(data['crash_date'])

# Extract the year from the crash_date column
data['year'] = data['crash_date'].dt.year

# Group the data by year and sum the death and injury counts for different categories
yearly_stats = data.groupby('year').agg({
    'motor_vehicle_death_count': 'sum',
    'bicycle_death_count': 'sum',
    'pedestrian_death_count': 'sum',
    'motorcycle_death_count': 'sum',
    'micromobility_death_count': 'sum',
    'sus_serious_injry_cnt': 'sum',
    'nonincap_injry_cnt': 'sum',
    'poss_injry_cnt': 'sum',
    'non_injry_cnt': 'sum',
    'unkn_injry_cnt': 'sum',
    'motor_vehicle_serious_injury_count': 'sum',
    'bicycle_serious_injury_count': 'sum',
    'pedestrian_serious_injury_count': 'sum',
    'motorcycle_serious_injury_count': 'sum',
    'micromobility_serious_injury_count': 'sum'
}).reset_index()

# Plot the histogram
plt.figure(figsize=(16, 10))

# Plot death counts
#plt.bar(yearly_stats['year'] - 0.4, yearly_stats['death_cnt'], width=0.1, label='Total Deaths')
#plt.bar(yearly_stats['year'] - 0.3, yearly_stats['apd_confirmed_death_count'], width=0.1, label='APD Confirmed Deaths')
plt.bar(yearly_stats['year'] - 0.2, yearly_stats['motor_vehicle_death_count'], width=0.1, label='Motor Vehicle Deaths')
plt.bar(yearly_stats['year'] - 0.1, yearly_stats['bicycle_death_count'], width=0.1, label='Bicycle Deaths')
plt.bar(yearly_stats['year'], yearly_stats['pedestrian_death_count'], width=0.1, label='Pedestrian Deaths')
plt.bar(yearly_stats['year'] + 0.1, yearly_stats['motorcycle_death_count'], width=0.1, label='Motorcycle Deaths')
plt.bar(yearly_stats['year'] + 0.2, yearly_stats['micromobility_death_count'], width=0.1, label='Micromobility Deaths')

# Plot injury counts
plt.bar(yearly_stats['year'] + 0.3, yearly_stats['sus_serious_injry_cnt'], width=0.1, label='Suspected Serious Injuries')
plt.bar(yearly_stats['year'] + 0.4, yearly_stats['nonincap_injry_cnt'], width=0.1, label='Non-incapacitating Injuries')
plt.bar(yearly_stats['year'] + 0.5, yearly_stats['poss_injry_cnt'], width=0.1, label='Possible Injuries')
plt.bar(yearly_stats['year'] + 0.6, yearly_stats['non_injry_cnt'], width=0.1, label='Non-injury Incidents')
plt.bar(yearly_stats['year'] + 0.7, yearly_stats['unkn_injry_cnt'], width=0.1, label='Unknown Injuries')
#plt.bar(yearly_stats['year'] + 0.8, yearly_stats['tot_injry_cnt'], width=0.1, label='Total Injuries')
plt.bar(yearly_stats['year'] + 0.9, yearly_stats['motor_vehicle_serious_injury_count'], width=0.1, label='Motor Vehicle Serious Injuries')
plt.bar(yearly_stats['year'] + 1.0, yearly_stats['bicycle_serious_injury_count'], width=0.1, label='Bicycle Serious Injuries')
plt.bar(yearly_stats['year'] + 1.1, yearly_stats['pedestrian_serious_injury_count'], width=0.1, label='Pedestrian Serious Injuries')
plt.bar(yearly_stats['year'] + 1.2, yearly_stats['motorcycle_serious_injury_count'], width=0.1, label='Motorcycle Serious Injuries')
plt.bar(yearly_stats['year'] + 1.3, yearly_stats['micromobility_serious_injury_count'], width=0.1, label='Micromobility Serious Injuries')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Number of Deaths and Injuries')
plt.title('Yearly Deaths and Injuries by Category')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

# Select relevant columns for the heatmap
heatmap_data = data[['latitude', 'longitude', 'tot_injry_cnt', 'death_cnt']]

# Compute the correlation matrix
corr_matrix = heatmap_data.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Heatmap of Latitude, Longitude, Injury Count, and Death Count')
plt.show()
```

```python

# Select relevant columns for the box plot
boxplot_data = data[['sus_serious_injry_cnt', 'nonincap_injry_cnt', 'poss_injry_cnt', 'non_injry_cnt', 'tot_injry_cnt', 'death_cnt']]

# Plot the box plot
plt.figure(figsize=(12, 8))
boxplot_data.boxplot()
plt.title('Box-and-Whisker Plot of Injury and Death Counts')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Select relevant columns for analysis
selected_columns = [
    'sus_serious_injry_cnt', 'nonincap_injry_cnt', 'poss_injry_cnt',
    'non_injry_cnt', 'tot_injry_cnt', 'death_cnt'
]

# Calculate correlation
correlation_matrix = data[selected_columns].corr()

# Calculate mean
mean_values = data[selected_columns].mean()

# Calculate median
median_values = data[selected_columns].median()

# Calculate standard deviation
std_dev_values = data[selected_columns].std()

print("Correlation Matrix:\n", correlation_matrix)
print("\nMean Values:\n", mean_values)
print("\nMedian Values:\n", median_values)
print("\nStandard Deviation Values:\n", std_dev_values)
```
